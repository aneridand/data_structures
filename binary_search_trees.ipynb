```python
from dtreeviz.trees import *
from lolviz import *
```

```python
class TreeNode():
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
```

```python
class BinarySearchTree():
    def __init__(self, root: TreeNode):
        self.root = root
    
    def search(self, root: TreeNode, x):
        if root is None:
            return None
        
        if x < root.value:
            return self.search(root.left, x)
        
        if x > root.value:
            return self.search(root.right, x)
        
        return root
    
    def add(self, root: TreeNode, x):
        if root is None:
            return  TreeNode(x)
        
        if x < root.value:
            root.left = self.add(root.left, x)
        
        else:
            root.right = self.add(root.right, x)
    
        return root
```

```python
root = TreeNode(7)
left = TreeNode(4)
right = TreeNode(9)

root.left = left
root.right = right
```

```python
treeviz(root)
```

```python
binary_search_tree = BinarySearchTree(root)
```

```python
binary_search_tree.search(root,9)
```

```python
root.left.left = TreeNode(1)
root.left.right = TreeNode(5)
root.right.left = TreeNode(8)
root.right.right = TreeNode(12)
```

```python
treeviz(root)
```

```python
binary_search_tree.add(root, 15)
```
